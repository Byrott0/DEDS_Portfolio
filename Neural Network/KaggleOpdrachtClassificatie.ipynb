#  Pandas Classification model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
```

# Data inlezen

```python
dataset = pd.read_csv('Supplement_Sales_Dataset.csv')
dataset
```

```python
dataset['Date'] = dataset['Date'].astype('datetime64[ns]')
dataset
```

Hierboven geprobeerd decision tree te maken door one hot encoding van Date naar datum te converten, maar het lukt nog steeds niet dus ik snij hem

# data voor x en y waardes weergeven

```python

X = dataset.drop(['Category', 'Date'], axis=1)
y = dataset['Category']
X
```

# one hot encoding

```python
# convert the string types data columns

X = pd.get_dummies(X, columns=['Product Name', 'Location', 'Platform'])
X
```

# trainen van data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.15,
    random_state=42)
```

```python
list(X.columns)
```

# max depth 2

```python
dectree = DecisionTreeClassifier(max_depth = 2)
dectree = dectree.fit(X_train, y_train)
tree.plot_tree(dectree, feature_names = X.columns)
plt.show()
```

```python
voorspelling = pd.DataFrame(dectree.predict(X_test))
voorspelling = voorspelling.rename(columns={0: 'voorspelde_Category'})
model_results_frame = pd.concat([y_test.reset_index()['Category'], voorspelling], axis = 1)
model_results_frame
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Haal de unieke labels op die voorkomen in echte en voorspelde data
labels = sorted(y_test.unique())

# Maak de matrix
conf_matrix = confusion_matrix(model_results_frame['Category'], model_results_frame['voorspelde_Category'], labels=labels)

# Visualiseer hem netjes
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=labels)
cm_display.plot(xticks_rotation=45)
plt.tight_layout()
plt.show()
```

```python
metrics.accuracy_score(model_results_frame['Category'], model_results_frame['voorspelde_Category'])
```
